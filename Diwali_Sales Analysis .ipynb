```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender', hue='Gender', legend=False, width=0.7, data = df)

for bars in ax.containers:
    ax.bar_label(bars)
ax.set_title('Distribution by Gender', fontweight='bold',color='navy')
ax.set_xlabel('Gender', fontsize=12,color='navy')
ax.set_ylabel('Count', fontsize=12,color='navy')
plt.tight_layout()
plt.show()
```

```python
# plotting a bar chart for gender vs total amount

import matplotlib.ticker as ticker
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
fig, ax = plt.subplots(figsize=(7,5))
def millions_formatter(x, pos):
    return f'{x/1000000:.1f}M'
ax.yaxis.set_major_formatter(ticker.FuncFormatter(millions_formatter))
ax = sns.barplot(x = 'Gender', hue='Gender', y= 'Amount' ,width=0.7, data = sales_gen)
ax.set_title('Total Amount by Gender', fontweight='bold',color='navy')
ax.set_xlabel('Gender', fontsize=12,color='navy')
ax.set_ylabel('Total Amount (millions)', fontsize=12,color='navy')
plt.tight_layout()
plt.show()
```

```python
# plotting a Stack chart for Age Group and it's count
age_order = ['0-17', '18-25', '26-35', '36-45', '46-50', '51-55', '55+']

ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender', order= age_order)
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_title('Count by Age Group', fontweight='bold',color='navy')
ax.set_xlabel('Age Group', fontsize=12,color='navy')
ax.set_ylabel('Count', fontsize=12,color='navy')
plt.show()
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
fig, ax = plt.subplots(figsize=(6,4))
def millions_formatter(x, pos):
    return f'{x/1000000:.1f}M'
ax.yaxis.set_major_formatter(ticker.FuncFormatter(millions_formatter))
ax=sns.barplot(x = 'Age Group', hue= 'Age Group', y= 'Amount' ,data = sales_age)
ax.set_title('Total Amount by Age Group', fontweight='bold',color='navy')
ax.set_xlabel('Age Group', fontsize=12,color='navy')
ax.set_ylabel( 'Amount', fontsize=12,color='navy')
plt.tight_layout()
plt.show()

#*From graphs we can see that most of the buyers are of age group between 26-35 yrs female*
```

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(8,3)})
ax= sns.barplot(data = sales_state, y = 'State', hue = 'State', x= 'Orders')
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_title('Number of Orders by States', fontsize=14, fontweight='bold',color='navy')
ax.set_ylabel('States', fontsize=12,color='navy')
ax.set_xlabel('Orders', fontsize=12,color='navy')
plt.tight_layout()
plt.show()
#*From graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*
```

```python
# Marital Status 

ax = sns.countplot(data = df, x = 'Marital_Status', hue = 'Marital_Status', width=0.7, legend=False)

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_title('Marital Status Count', fontsize=14, fontweight='bold',color='navy')
ax.set_xlabel('Marital Status', fontsize=12,color='navy')
ax.set_ylabel('Count', fontsize=12,color='navy')
plt.tight_layout()
plt.show()
```

```python
# plotting a Stack chart for Purchase Amount by Marital Status
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(7,5)})
ax = sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
ax.set_title('Purchase Amount by Marital Status', fontsize=14, fontweight='bold',color='navy')
ax.set_xlabel('Marital Status', fontsize=12,color='navy')
ax.set_ylabel('Amount', fontsize=12,color='navy')
plt.show()
# *From graphs we can see that most of the buyers are married (women) and they have high purchasing power*
```

```python
#plotting a chart for Occupation Frequency

sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation', hue='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
ax.set_title('Occupation Frequency', fontsize=14, fontweight='bold',color='navy')
ax.set_xlabel('Occupation', fontsize=12,color='navy')
ax.set_ylabel('Count', fontsize=12,color='navy')
plt.show()
```

```python
##plotting a chart for Spending Across Different Occupations

sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
ax=sns.barplot(data = sales_state, x = 'Occupation',hue='Occupation', y= 'Amount')
ax.set_title('Spending Across Different Occupations', fontsize=14, fontweight='bold',color='navy')
ax.set_xlabel('Occupation', fontsize=12,color='navy')
ax.set_ylabel('Amount', fontsize=12,color='navy')
plt.show()

#*From  graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*
```

```python
# plotting a chart for Distribution of Products by Category

sns.set(rc={'figure.figsize':(8,5)})
ax = sns.countplot(data = df, y = 'Product_Category', hue = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
ax.set_title('Distribution of Products by Category', fontsize=14, fontweight='bold',color='navy')
ax.set_ylabel('Product_Category', fontsize=12,color='navy')
ax.set_xlabel('Count', fontsize=12,color='navy')
plt.show()
```

```python
#plotting a chart for Spending by Product Category

sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',hue = 'Product_Category', y= 'Amount')
ax.set_title('Spending by Product Category"', fontsize=14, fontweight='bold',color='navy')
ax.set_ylabel('Product_Category', fontsize=12,color='navy')
ax.set_xlabel('Count', fontsize=12,color='navy')
plt.show()

#*From graphs we can see that most of the sold products are from Food, Clothing and Electronics category*
```

```python
# Method 1: plotting a chart for Orders by Product ID

sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(data = sales_state, y = 'Product_ID',hue = 'Product_ID', x= 'Orders')
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_title('Orders by Product ID', fontsize=14, fontweight='bold',color='navy')
ax.set_xlabel('Orders', fontsize=12,color='navy')
ax.set_ylabel('Product ID', fontsize=12,color='navy')
plt.show()
# top 10 most sold products (same thing as above)
```

```python
# Method 2: top 10 most sold products (same thing as above)
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(data = sales_state, x = 'Product_ID',hue = 'Product_ID', y= 'Orders')
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_title('top 10 most sold products', fontsize=14, fontweight='bold',color='navy')
ax.set_xlabel('Orders', fontsize=12,color='navy')
ax.set_ylabel('Product ID', fontsize=12,color='navy')
plt.show()
```
